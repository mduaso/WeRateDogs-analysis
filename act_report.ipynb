# Analysis Report - WeRateDogs

After assessing and cleaning the tweets dataset, we want to analyze the data to get some insights. We are going to look into 3 areas of interest to further understand them:

- Dog image ratings
- Number of retweets and favorite counts
- Dog breeds 

### 1st insight - Dog image ratings

Let's take a look at the distribution of dog image ratings. We'll use a histogram to plot the variable *rating*, since it is a continuous variable: 

What we see above is a __left skewed distribution__, with most values falling in the 0.75-1.25 range (very good ratings generally!) and some outliers with an extreme value of 2.0.

On any traditional rating system, we would expect values to be distributed between 0 and 1, but we're not seeing that in this case. For context here, let's recall that the original *WeRateDogs*'s ratings (from 0-10) didn't actually enforce an upper bound of 10, allowing for values as high as users want. 

Because of that, we decided to set an artificial boundary on a rating of 20/10. Any rating above that is reduced to 2.0, reason why we have some outliers with that specific value.  

### 2nd insight - Number of retweets and favorite counts

Let's take a look at the variables *retweet count* and *favorite count*. Do they have any correlation? Take a look at the scatter plot for both variables:

We can observe a strong linear relation between *retweets* and *favorite count*. The correlation coefficient between these two variables is 0.927, which supports the observed correlation in the plot.

Also, even though most tweets fall on the lower left corner of the plot (small number of retweets and favorite counts), there are some outliers corresponding to extremely poopular tweets. One in particular (right-most point) has close to 80 thousand retweets and 160 thousand favorite counts.
